```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
list1=[[1,45],[2,48],[3,51],[4,37]]
```

```python
df=pd.DataFrame(list1)
df.columns=["house","price"]
df
```

```python
plt.plot(df.house,df.price)
plt.show()
```

```python
x=df.house
x=x.values.reshape(-1,1)
```

```python
y=df.price
y=y.values.reshape(-1,1)
```

```python
model= LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[5]])
```
